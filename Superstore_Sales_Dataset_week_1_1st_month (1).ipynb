**Importing Essential Libraries**

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

**Importing Dataset**

```python
df=pd.read_excel("/content/04-01-Financial Sample Data.xlsx")
```

```python
df
```

**Data Cleaning**

```python
df.dtypes
```

```python
df.isnull().sum()
```

**coverting date number into real date**

```python
df['RealDate'] = df['Date'].apply(lambda x: datetime(1900,1,1) + timedelta(days=x - 2))
```

```python
df.head()
```

**Drop unnecessary columns,  manufactiring price in not correct**

```python
df.drop('Menufacturing Cost', axis=1, inplace=True)
```

```python
df.drop('Manufacturing Price', axis=1, inplace=True)
```

**Creating new column**

```python
df['Manufacturing Price']=df.COGS/df['Units Sold']
```

```python
df.head()
```

**Reposition new column**

```python
manufacturing_price = df.pop('Manufacturing Price')
df.insert(5, 'Manufacturing Price', manufacturing_price)
```

```python
df.head()
```

```python
df.drop('Date', axis=1, inplace=True)
```

df.to_csv("financial_sample_week5.csv", index= False)
